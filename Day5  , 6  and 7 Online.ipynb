```python
import numpy as np
import matplotlib.pyplot as plt
print('importing of lib is completed')
```

```python
# sinewave
#A*sin(2*pi*f*t)
#A- amplitude
#f- frequency
#t- time , time samples
f=2 # hz
pi=3.142
t=np.arange(0,1,1/100)
Y=np.sin(2*pi*f*t)
plt.plot(t,Y)
plt.show()
```

```python
f=2 # hz
pi=3.142
t=np.arange(0,1,1/100)
Y=np.sin(2*pi*f*t)
Z=np.cos(2*pi*f*t)
YZ=Y+3*Z

plt.plot(t,Y,'r')
plt.show()
plt.plot(t,Z,'g*')
plt.show()
plt.plot(t,YZ,'c+:')
plt.show()
# all three plots plotted in seearate window
```

```python
plt.plot(t,Y,'r')
plt.plot(t,Z,'g*')
plt.plot(t,YZ,'c+:')
plt.show() # all above three plot overlapped on single window
```

```python
plt.plot(t,Y,'r',t,Z,'g*') # 2 plot in one plot function, we can have more than 2 also
plt.show()
```

# PANDAS
***What is Pandas: is a open source data analysis and manipulation tool, which is built on top of Python
- Pandas supports Reading data stored in various formats like, CSV, Excel sheets, SQL Data bases, HDF5 JSON, file format etc, we can also write the data back to above mentioned file formats using Pandas
- Data read from pandas lib is called dataFrame, usually dataframe can have both numerical as well as categorical(string) data in one file
- We can access the contetn stored in Dataframe just like numpy array indexing also has several easies option to access the contetn, we can also manipulate the content,
- we can use many methods of pandas to get more insight into dataframe, methods to check datatypes, statstical information(min, max, mean, std), methods to check null values, unique values etc
- we can join merge delete dataframes
- application almost all Datscience application uses pandas( Neuroscience , economics, stats, ads web analytics much more)
